# Check if Item Exists

```python
data = list(("apple","ball","cat","dog","eye"))

if "eye" in data:
    print("Word 'eye' found in data.")
```

# Change Item Value

```python
data = list(("apple","ball","cat","dog","eye"))

print(data)
data[1] = "banana"
print(data)
```

# Change a Range of Item Values

```python
data = list(("apple","ball","cat","dog","eye"))

print(data)
data[1:3] = ["banana","can"]
print(data)
```

```python
# If you insert more items than you replace, the new items will 
# be inserted where you specified, and the remaining items 
# will move accordingly.

data = ["apple","ball","cat"]

print(data) 

data[0:1] = ["aeroplane","banana"]
# here we should have added only one value
# but we added 2 which results in shifting of other data in list

print(data)
```

```python
'''
If you insert less items than you replace, the new items 
will be inserted where you specified, and the remaining 
items will move accordingly:
'''
# we 😉replace multiple value by replacing it with one value

data = ["apple","ball","cat","dog","elephant"]
print(data)

data[1:4] = ["replaced"]
#lets replace 1 ,2 ,3
print(data)
```

# Insert in List
To insert a list item at a specified index, use the insert() method.

```python
data = ["maya","xaya","kaya","kyakhaya"]
print(data)

data.insert(1,"kuchnahipaya")
print(data)
```

# Append Items in List

To add an item to the end of the list, use the append() method:

```python
data = ["maya","xaya","kaya","kyakhaya"]
print(data)

data.append("kuchnahipaya")
print(data)
```

# Extend List
To append elements from another list to the current list, use the extend() method.

```python
list1 = [1,2,3,4,5]
list2 = [5,6,7,8,9]

list1.extend(list2)
print(list1)
```

# Extend NOTE
The extend() method does not have to append lists, you can add any iterable object (tuples, sets, dictionaries etc.).

```python
thislist = ["apple", "banana", "cherry"]
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(thislist)
```

# List has further more Fetures :


Remove Specified Item

The remove() method removes the specified item.

```python
data = ["maya","xaya","kaya","kyakhaya"]

data.remove("maya")

print(data)
```

Remove Specified Index

The pop() method removes the specified index.

```python
data = ["maya","xaya","kaya","kyakhaya"]

data.pop(0)

print(data)
```

The del keyword removes the specified index or can also delete the list completely.

```python
data = ["maya","xaya","kaya","kyakhaya"]

del data[0]
print(data)

del data

# print(data) if we print then we get error
```

Clear the List
The clear() method empties the list.

The list still remains, but it has no content.

```python
data = ["maya","xaya","kaya","kyakhaya"]

print(data)

data.clear()

print(data)
```

# Loop Through a List

```python
data = ["maya","xaya","kaya","kyakhaya"]

for x in data:
    print(x)
```

Loop Through the Index Numbers
You can also loop through the list items by referring to their index number.

Use the range() and len() functions to create a suitable iterable.

```python
data = ["maya","xaya","kaya","kyakhaya"]

for x in range(len(data)):
    print(data[x])
```

```python
data = ["maya","xaya","kaya","kyakhaya"]
x = 0
while x < len(data):
    print(data[x])
    x+=1
```

```python
# copying data of list to another if words contains "a"

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
temp = []

for x in fruits:
    if "a" in x:
        temp.append(x)
        
print(temp)
```

# Python - Sort Lists
Sort List Alphanumerically
List objects have a sort() method that will sort the list alphanumerically, ascending, by default:

```python
data = ["maya","xaya","kaya","kyakhaya"]
print(data)

data.sort()#Alphabetically
print(data)

data.sort(reverse = True)#Descending
print(data)
```

```python
# Reverse the order of the list items:

thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

```python
# Copy a List

thislist = ["banana", "Orange", "Kiwi", "cherry"]
copied = thislist.copy()
print(copied)
```
